```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36"}
```

```python
url = f"https://finance.naver.com/sise/sise_group_detail.naver?type=upjong&no=304"
res = requests.get(url, headers = headers)
res.raise_for_status()
soup = BeautifulSoup(res.text, "lxml")
```

```python
soup.select('.type_5 > tbody > tr > td.number')[0].text.replace(',','') # 현재가
soup.select('.type_5 > tbody > tr > td.number')[1].select('span')[0].text.replace('\n','').replace('\t','').replace(',','') # 전일비
soup.select('.type_5 > tbody > tr > td.number')[2].select('span')[0].text.replace('\n','').replace('\t','') # 등락률
soup.select('.type_5 > tbody > tr > td.number')[3].text.replace(',','') # 매수호가
soup.select('.type_5 > tbody > tr > td.number')[4].text.replace(',','') # 매도호가
soup.select('.type_5 > tbody > tr > td.number')[5].text.replace(',','') # 거래량
soup.select('.type_5 > tbody > tr > td.number')[6].text.replace(',','') # 거래대금
soup.select('.type_5 > tbody > tr > td.number')[7].text.replace(',','') # 전일거래량
```

```python
for i in range(0, len(soup.select('.type_5 > tbody > tr > td.name > div.name_area > a'))):
    print(soup.select('.type_5 > tbody > tr')[i].select('div.name_area > a')[0].text)
```

```python
전체종목=[]
for i in range(0, len(soup.select('.type_5 > tbody > tr > td.name > div.name_area > a'))):
    전체종목.append(soup.select('.type_5 > tbody > tr')[i].select('div.name_area > a')[0].text)
print(전체종목)
```

```python
종목별현재가=[]
for i in range(0, len(soup.select('.type_5 > tbody > tr > td.name > div.name_area > a'))):
    종목별현재가.append(soup.select('.type_5 > tbody > tr')[i].select('td.number')[0].text.replace(',',''))
print(종목별현재가)
```

```python
종목별전일거래량=[]
for i in range(0, len(soup.select('.type_5 > tbody > tr > td.name > div.name_area > a'))):
    종목별전일거래량.append(soup.select('.type_5 > tbody > tr')[i].select('td.number')[7].text.replace(',',''))
print(종목별전일거래량)
```
