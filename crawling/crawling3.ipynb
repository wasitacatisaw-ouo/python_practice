## 미세먼지 등

```python
from urllib.request import urlopen, Request
import urllib
import bs4
```

```python
location = '하남'
en_loc = urllib.parse.quote(location + ' 날씨') # ------- 인코딩
url = 'https://search.naver.com/search.naver?ie=utf8&query=' + en_loc
req=Request(url)
page=urlopen(req)
html=page.read()
soup=bs4.BeautifulSoup(html, 'lxml') # ------- dom으로 변환, lxml을 통하여 의미있는 HTML문서로 변환시킨다
```

```python
temp=soup.find('div', class_='temperature_text').find('strong').get_text()
tm=temp[5::]
final_tm=tm[:size-1] # ------ 도 특수문자 제거
cast=soup.find('div', 'weather_main').get_text()
print('현재 '+location+'의 날씨는'+cast+'이며, 기온은 '+final_tm+'도 입니다')
```

```python
find_dust = soup.find('div', 'weather_info')
find_dust
```

```python
find_dust = soup.find('div', 'weather_info').find_all('li', class_='item_today level1')[0].find('span').get_text()
find_dust
```

```python
find_ultra_dust = soup.find('div', 'weather_info').find_all('li', class_='item_today level2')[0].find('span').get_text()
find_ultra_dust
```

```python
find_sunset = soup.find('div', 'weather_info').find_all('li', class_='item_today type_sun')[0].find('span', class_='txt').get_text()
find_sunset
```

```python
print('현재 '+location+'의 날씨는'+cast+'이며, 기온은 '+final_tm+'도 입니다\n미세먼지는 '
      +find_dust+'이며, 초미세먼지는 '+find_ultra_dust+', 일몰시간은 '+find_sunset+'입니다')
```
