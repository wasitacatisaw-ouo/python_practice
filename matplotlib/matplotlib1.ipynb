# visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 4, 9, 16, 25])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot([1, 2, 4, 9, 16, 25])
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,50,0,100])
plt.show()
```

```python
plt.plot([1, 2, 4, 9, 16, 25],'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,50,0,100])
plt.show()
```

```python
plt.plot([1, 2, 4, 9, 16, 25],'bo') # bo  . yo ro 
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,50,0,100])
plt.show()
```

```python
plt.plot([1, 2, 4, 9, 16, 25],color='springgreen')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,7,0,30])
plt.show()
```

```python
plt.plot([1, 2, 4, 9, 16, 25], marker='o', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,8,0,30])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(0,2,0.2)
print(a,type(a))# [0.  0.2 0.4 0.6 0.8 1.  1.2 1.4 1.6 1.8] <class 'numpy.ndarray'>
plt.plot(a,a,'r--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(0,2,0.2)
print(a,type(a))# [0.  0.2 0.4 0.6 0.8 1.  1.2 1.4 1.6 1.8] <class 'numpy.ndarray'>
plt.plot(a,a,'r--',a,a**2,'bo')
plt.show()
```
