### Matplotlib

```python
import matplotlib.pyplot as plt
plt.plot([1,5,8,9],marker='*')
plt.title('Matplotlib Title')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,5,8,9],[3,6,9,12],marker='*') #(1,3)(5,6),(8,)
plt.title('Matplotlib Title')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,5,8,9],[3,6,9,12],[3,7,12,20],[2,7,10,15],marker='o')
plt.title('Matplotlib Title')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,5,8,9],[3,6,9,12],[3,7,12,20],[2,7,10,15],marker='*',linewidth=20)
plt.title('Matplotlib Title')
plt.axis([0,15,0,15])
plt.show()
```

```python
dic=dict()
print(dic,type(dic))
dic['hour']=[1,2,3,4,5,6]
dic['score']=[10,27,38,46,80,99]
print(dic)
```

```python
plt.plot('hour','score',data=dic, marker='^')
plt.xlabel('hour')
plt.ylabel('score')
plt.axis([1,7,10,100])
plt.show()
```
