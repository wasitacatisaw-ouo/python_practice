```python
frequency_d = {
    'python':1, 
    'big data':10,
    'crawling':4,
    'analysis':4,
    'visualization':4,
    'machine learning':5,
    'deep learning':6
}
wc = WordCloud(width=1000, height=600, background_color="white", random_state=10, font_path=r'c:\Windows\Fonts\malgun.ttf')
plt.imshow(wc.generate_from_frequencies(frequency_d))
plt.axis("off")
plt.show()
```
