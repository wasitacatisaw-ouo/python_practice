```python
키 = [170,180,175,160]
신발 = [260,270,265,255]
# 신발 = 키*a + b
```

```python
a = tf.Variable(0.1) # ---- w1
b = tf.Variable(0.1) # ---- w0

def 손실함수(): # ---- 오차의 제곱
    예측값 = 키*a+b
    return tf.square(260-예측값)

opt = tf.keras.optimizers.Adam(learning_rate=0.1) # ---- optimize 객체

for i in range(300):
    opt.minimize(손실함수, var_list=[a,b]) # ---- 경사하강법 통해 최적의 가중치 찾기
    print('a=', a.numpy())
    print('b=', b.numpy())
```

```python
train_x=[1,2,3,4,5,6,7]
train_y=[3,5,7,9,11,13,15]
a=tf.Variable(0.1)
b=tf.Variable(0.1)
```

```python
def 손실함수(a,b):
    예측_y=train_x*a+b
    return tf.keras.losses.mse(train_y,예측_y) # ---- 평균제곱오차
opt=tf.keras.optimizers.Adam(learning_rate=0.1)
```

```python
for i in range(500):
    opt.minimize(lambda:손실함수(a,b), var_list=[a,b]) # ---- opt.minimize()에는 함수 이름(주소값)만 가능하다 = lambda사용
    print('a=',a.numpy()) # ---- for문이 반복될수록 a는 2에 수렴
    print('b=',b.numpy()) # ---- for문이 반복될수록 b는 1에 수렴
```
