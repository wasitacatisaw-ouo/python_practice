```python
xdata = []
for i, rows in data.iterrows():
    xdata.append([rows['gre'], rows['gpa'], rows['rank']])
data.fillna(100)
```

```python
model=tf.keras.models.Sequential([ # ---- 노드의 갯수는 결과가 잘 나올 때 까지 실험으로 파악한다
    tf.keras.layers.Dense(64,activation='tanh'),
    tf.keras.layers.Dense(128,activation='tanh'),
    tf.keras.layers.Dense(1,activation='sigmoid'), # ---- 마지막 레이어 1,0 sigmoid는 0~1사이값 
    
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(np.array(xdata),np.array(ydata),epochs=1000)
예측값=model.predict([[750,3.70,3],[400,2.2,1]])
print(예측값)
```
