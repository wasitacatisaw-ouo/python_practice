```python
import numpy as np
import matplotlib.pyplot as plt

def pprint(arr):
    print('type:{}'.format(type(arr)))
    print("shape:{},dimension:{},dtype:{}".format(arr.shape,arr.ndim,arr.dtype))
    print("Array's Data:\n",arr)
    
a=np.linspace(0,1000,20)  #요소 개수를 기준으로 균등 간격의 배열을 생성
pprint(a)
# [   0.           52.63157895  105.26315789  157.89473684  210.52631579
#   263.15789474  315.78947368  368.42105263  421.05263158  473.68421053
#   526.31578947  578.94736842  631.57894737  684.21052632  736.84210526
#   789.47368421  842.10526316  894.73684211  947.36842105 1000. 
# type:<class 'numpy.ndarray'>
# shape:(20,),dimension:1,dtype:float64
# Array's Data:
```

```python
a=np.linspace(0,100,50)  #요소 개수를 기준으로 균등 간격의 배열을 생성 0에서 100이 범위 50은 요소 갯수 
pprint(a)
plt.plot(a,'o')
plt.show()
```

* 슬라이싱, 인덱싱

```python
array1 = np.arange(start=1, stop=10)
print('array1:', array1)
array3 = array1[0:3]
print('array3:', array3)
print(type(array3))
```

```python
array1 = np.arange(start=1, stop=20)
print('array1:', array1)
array3 = array1[-10:-1]
print('array3:', array3)
print(type(array3))
```

```python
array1d=np.arange(1,10)
print('array1d:', array1d)
array2d=array1d.reshape(3,3)
print('array2d:\n', array2d)

print(array2d[0]) # ---------- 0행 전부 -> [1 2 3]
print(array2d[1]) # ---------- 1행 전부 -> [4 5 6]
print(array2d[0:3]) # ---------- 0~3행
print(array2d[0:2,2]) # ---------- 0~2행에서 3열만 -> [3 6]
print(array2d[0:2,0]) # ---------- 0~2행에서 1열만 -> [1 4]
```

```python
print(array2d[0:2,0:2]) # -- 0~2행, 0~2열 -> [[1 2]
#                                             [4 5]]

print(array2d[0:2,0:3]) # -- 0~2행, 0~3열 -> [[1 2 3]
#                                             [4 5 6]]

print(array2d[1:2,0:2]) # -- 1~2행, 0~2열 -> [[4 5]]
#

print(array2d[1:3,0:3]) # -- 1~3행, 0~3열 -> [[4 5 6]
#                                             [7 8 9]]
```

```python
print('array2d[0] shape:', array2d[0].shape)

#  array2d[0][0]은 array2d의 요소이므로 shape 안나온다
print('array2d[0][0] shape:', array2d[0][0].shape)
print('array2d[0][1] shape:', array2d[0][1].shape)
```

* 팬시인덱싱

```python
array=np.arange(1,10)
print(array)
array2=array.reshape(3,3)
print(array2)
```

```python
array3=array2[[0,1]]
array4=array2[0,1]
print(array3, array3.shape)
print(array4, array4.shape)
```

```python
array3=array2[[0,2]] #  ---- 0행, 2행 -> [[1 2 3]
#                                         [7 8 9]]
array4=array2[2]
print(array3, array3.shape)
print(array4, array4.shape)
```

```python
array3=array2[[0]]
print(array3, array3.shape)
```

```python
array=np.arange(1,26)
print(array)
array2=array.reshape(5,5)
print(array2)
```

```python
array3=array2[[0,1]]
array4=array2[0,1]
print(array3, array3.shape)
print(array4, array4.shape)
```

```python
array3=array2[[1,3]] #  ---- 1행, 3행 -> [[ 6  7  8  9 10]
#                                         [16 17 18 19 20]]
array4=array2[2]
print(array3, array3.shape)
print(array4, array4.shape)
```

```python
array3=array2[[0,1],2] #  ---- 0행, 1행의  2열 -> [3 8]

print(array3, array3.shape)
print(array3.tolist())
```

```python
array3=array2[2:4] #  2 ~ 3행
array4=array2[2,4] #  2, 3행

print(array3, array3.shape)
print(array3.tolist())
print(array4, array4.shape)
print(array4.tolist())
```

```python
array3=array2[[2]] #  2행
array4=array2[[2,4]] #  2행, 4행

print(array3, array3.shape)
print(array3.tolist())
print(array4, array4.shape)
print(array4.tolist())
```

```python
array3=array2[[0,2],4] #  0행과 2행의 4열
array4=array2[[0,1],[1,2]] #  0행 1열, 1행 2열

print(array3, array3.shape)
print(array3.tolist())
print(array4, array4.shape)
print(array4.tolist())
```
