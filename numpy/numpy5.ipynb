```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pprint(arr):
    print("type:{}".format(type(arr)))
    print("shape: {}, dimension: {}, dtype:{}".format(arr.shape, arr.ndim, arr.dtype))
    print("Array's Data:\n", arr)
```

```python
# 정균분포도  normal(loc(분포평균,(scale 표준편차),size(갯수))
a=np.random.normal(size=50)
#print(a)
a=np.random.normal(0,1,1000)#1000
#pprint(a)
plt.hist(a,bins=10)# 뿌려지는 구간 갯수
```

```python
data = np.random.randint(-100, 100, 1000)# -100에서 +100이 범위 1000은 난수 갯수 
plt.hist(data, bins=10)
plt.show()
```

```python
data = np.random.random(10)# 1차원 구조에 0~1사이의 난수값을 생성 
print(data)
plt.hist(data, bins=10)
plt.show()
```

```python
# seed 값을 설정하여 아래에서 난수가 재연 가능하도록 함
np.random.seed(100)
print(np.random.random((2, 3)))# random 0~1 실수 난수발생  
print(np.random.randint(0, 4, (2, 3)))
```
