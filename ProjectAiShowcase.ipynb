# <hr>KNN 예측

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

userDataDf = pd.read_csv('testUserData.csv').dropna()

X = userDataDf[['Day', 'TimeSlot', 'PlayTime', 'PreferedGame', 'Gender', 'Age']]
y = userDataDf['UserType']

X = pd.get_dummies(X, columns=['Day', 'TimeSlot', 'PreferedGame', 'Gender'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print(y_pred)
print('정확도:', accuracy)
```

# <hr> KNN 예측값 코사인유사도

```python
from sklearn.metrics.pairwise import cosine_similarity

weights = {'Day': 0.3, 'TimeSlot': 0.3, 'PlayTime': 0.2, 
           'PreferedGame': 0.5, 'Gender': 0.1, 'Age': 0.4, 'UserType': 0.7}

user_vectors = []
for index, row in userDataDf.iterrows():
    vector = []
    for features in ['Day', 'TimeSlot', 'PlayTime', 'PreferedGame', 'Gender', 'Age', 'UserType']:
        vector.append(row[features] * weights[features])
    user_vectors.append(vector)

vector_similarity = cosine_similarity(user_vectors)

print(vector_similarity[0])
```

# <hr> KNN 예측값 유클리드거리

```python
from sklearn.metrics.pairwise import euclidean_distances

weights = {'Day': 0.3, 'TimeSlot': 0.3, 'PlayTime': 0.2, 
           'PreferedGame': 0.5, 'Gender': 0.1, 'Age': 0.4, 'UserType': 0.7}

user_vectors = []
for index, row in userDataDf.iterrows():
    vector = []
    for features in ['Day', 'TimeSlot', 'PlayTime', 'PreferedGame', 'Gender', 'Age', 'UserType']:
        vector.append(row[features] * weights[features])
    user_vectors.append(vector)

vector_distance = euclidean_distances(user_vectors)

print(vector_distance)
```

# <hr> Decision Tree Classifier 예측

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd

userDataDf = pd.read_csv('testUserData.csv').dropna()

dummy_cols = ['Day', 'TimeSlot', 'PreferedGame', 'Gender']
userDataDf = pd.get_dummies(userDataDf, columns=dummy_cols)

X = userDataDf.drop(['UserType', 'MembersNo'], axis=1)
y = userDataDf['UserType']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)

#print(classification_report(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)

print('정확도:', accuracy)
```

# <hr> Decision Tree Classifier 예측값 코사인유사도

```python
from sklearn.metrics.pairwise import cosine_similarity

# 1번 유저와 다른 유저들 간의 코사인 유사도 추출
user1 = X_test.iloc[0].values.reshape(1,-1)  # 1번 유저 데이터 추출
other_users = X_test.drop(X_test.index[0])  # 1번 유저를 제외한 나머지 유저 데이터 추출

# 코사인 유사도 추출
similarities = cosine_similarity(user1, other_users)

# 결과 출력
print(similarities)
```
